## Distribution of data based on the quartiles and potential outliers

```python
plt.figure(figsize=(8, 5))
ax = sns.boxplot(data=df[numeric_columns.columns])
plt.xticks(rotation=45)
plt.title('Boxplot of Numeric Columns')
plt.xlabel('Columns')
plt.ylabel('Values of each column')
plt.grid(True)

mean_values = df[numeric_columns.columns].mean().values
plt.plot(range(len(mean_values)), mean_values, marker='o', markerfacecolor='white', markersize=6, markeredgecolor='black', linestyle='', label='Mean')

plt.legend()
plt.show()
```

## Detect outliers using Interquartile Range

```python
Q1 = df['total_price'].quantile(0.25)
Q3 = df['total_price'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['total_price'] < Q1 - 1.5 * IQR) | (df['total_price'] > Q3 + 1.5 * IQR)]

outliers
```

## Average Unit Price and Revenue by Category

```python
category_analysis = df.groupby('pizza_category').agg(
    average_unit_price=('unit_price', 'mean'),
    revenue_per_category=('unit_price', lambda x: (x * df['quantity']).sum())
).sort_values(by='revenue_per_category', ascending=False)
print("Average Unit Price and Revenue by Category:\n", category_analysis)
```

## Revenue by Pizza Category

```python
category_analysis['revenue_per_category'].plot(kind='bar', color='skyblue')
plt.xlabel('Pizza Category')
plt.ylabel('Revenue')
plt.title('Revenue by Pizza Category in ')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Average Unit Price and Revenue of Pizza by Size

```python
size_analysis = df.groupby('pizza_size').agg(
    average_unit_price=('unit_price', 'mean'),
    revenue_per_size=('unit_price', lambda x: (x * df['quantity']).sum())
).sort_values(by='revenue_per_size', ascending=False)
print("Average Unit Price and Revenue by Size:\n", size_analysis)
```

## Revenue by Pizza Size Plot

```python
size_analysis['revenue_per_size'].plot(kind='bar', color='blue')
plt.xlabel('Pizza Size')
plt.ylabel('Revenue')
plt.title('Revenue by Pizza Size in ')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Best Selling Pizza 

```python
best_selling_pizza_analysis = df.groupby('pizza_name').agg(
    average_unit_price=('unit_price', 'mean'),
    revenue_per_pizza=('unit_price', lambda x: (x * df['quantity']).sum())
).nlargest(8, 'revenue_per_pizza')
print("Average Unit Price and Revenue of Top 8 Pizzas:\n", best_selling_pizza_analysis)
```

Best Selling Pizzas Revenue

```python
best_selling_pizza_analysis['revenue_per_pizza'].plot(kind='bar', color='green')
plt.xlabel('Pizza Name')
plt.ylabel('Revenue')
plt.title('Revenue by Best Selling Pizzas in ')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Busiest Time of the Day

## The Most Used Ingredients

```python
ingredients_col = df["pizza_ingredients"]
ingredient_count = {}
for ingredients in ingredients_col:
  ingredient_list = ingredients.split(",")
  for ingredient in ingredient_list:
    ingredient = ingredient.strip()
    if ingredient not in ingredient_count:
      ingredient_count[ingredient] = 1
    else:
      ingredient_count[ingredient] +=1
sorted_ingredients = dict(sorted(ingredient_count.items(), key=lambda x:x[1], reverse = True))
print(sorted_ingredients)
print(len(ingredient_count.keys()))
```

```python
ingredient_name = list(sorted_ingredients.keys())
count = list(sorted_ingredients.values())
top_ing = ingredient_name[:20]
top_count = count[:20] 
px.bar(x=top_ing, y=top_count, color = top_count, color_continuous_scale="aggrnyl_r", text = top_count,
       title = "Most Used Ingredients").update_layout(title_x=0.5, xaxis_title = "Ingredients", yaxis_title="Count",
                                                      title_font = dict(size=25,color="green"))
```

##### Garlic and Tomatoes are the most popular ingredients/ pizza toppings. Followed by Red Onions and Red Peppers. Knowing this can help optimise stock levels, reduce wastage and increase profits.
